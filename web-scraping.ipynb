```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.50pros.com/fortune500'
page = requests.get(url=url)
soup = BeautifulSoup(page.text, 'html')
```

```python
# now find the elements you want
element = soup.find_all('iframe')
element
len(element)
```

```python
iframe_page = requests.get(url='https://sheet2site.com/api/v3/index.php?key=1QjkZQqZiBPNOiEgsR0PlARV275MgdYt0qY00x_cv1fs&g=1&e=1&e=1')
chili = BeautifulSoup(iframe_page.text,'html')

chili
```

```python
table = chili.find_all('table')[0]
# table
headers = table.findAll('th')
# len(headers)
# 

columns = [header.text.strip() for header in headers]
# columns
headers
```

```python
columns
```

```python
company_df = pd.DataFrame(columns=columns)
```

```python
tr = table.findAll('tr')[1:]
# rows = [td.find_all('td') for td in tr]
for td in tr:
   rows = td.find_all('td')
   row_data = [info.text.strip() for info in rows]
   length = len(company_df)
   company_df.loc[length] = row_data
```

```python
headers = table.find_all('th')
global_columns = [header.text.strip() for header in headers]
# global_columns
global_df = pd.DataFrame(columns=global_columns)
```

```python
tr = table.find_all('tr')[1:]
for r in tr:
    row = r.find_all('td')
    row_data = [info.text.strip() for info in row]
    length = len(global_df)
    global_df.loc[length] = row_data
```
